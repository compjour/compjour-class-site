```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
data = [10, 20, 15, 30, 25]
```

## A simple plot

Matplotlib docs: [plot()](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot).

The `plot()` function will draw a line chart. When using iPython or an iPython notebook, invoking `plot()` (or other charting functions, such as [bar()](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.bar) and [pie()](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.pie)) will display the resulting chart. 


```python
plt.plot(data)
```

## Adding a title

Matplotlib docs: [suptitle()](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.suptitle).

The `suptitle()` function is used to display a title for the _entire_ chart, which is more apparent when creating charts that consist of several subplots.


```python
plt.suptitle("Main Title", fontsize = 30)
plt.plot(data)
```

## Add axis labels

Matplotlib docs: [xlabel()](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.xlabel) and [ylabel()](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.ylabel).


Labels are instances of [matplotlib.text.Text](http://matplotlib.org/users/text_props.html).

Colors can be represented as a __tuple__ of RGB colors or as a __string__ for a predefined name, e.g. `'blue'` and `'gold'` (see the [full list of names](http://matplotlib.org/examples/color/named_colors.html))



```python
plt.xlabel('X Factor', color = (0.9, 0, 0.9), fontsize = 20)
plt.ylabel('Y Combos', color = 'Chocolate', fontsize = 16, weight = 'bold', fontname = 'Comic Sans MS')
plt.plot(data)
```

## Formatting axis ticks

Matplotlib docs: [xticks()](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.xticks) and [yticks()](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.yticks).

[ticks_and_spines example code: ticklabels_demo_rotation.py](http://matplotlib.org/examples/ticks_and_spines/ticklabels_demo_rotation.html)


```python
plt.xticks(rotation = 'vertical')
plt.yticks(rotation = -45)
plt.plot(data)
```

## Labeling axis ticks

```python
plt.yticks([15, 30], ['Meh', 'Whoa!'])
plt.xticks([0, 1, 2, 3, 4], ['Zero', 'Un', 'Deux', 'Tres', 'Cinco'])
plt.plot(data)
```

## Setting the axis limit

Matplotlib docs: [xlim()](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.xlim) and [ylim()](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.ylim).


By default, the lower and upper bounds are set based on the max and min of the data. The `xlim()` and `ylim()` functions can be used to set the beginning and/or end of an axis's range.


```python
plt.xlim(xmax = 10)
plt.ylim([0, 50]) 
plt.plot(data)
```

## Using stylesheets

Matplotlib docs: [Customizing plots with style sheets](http://matplotlib.org/users/style_sheets.html)

The tedious process of setting up visual styles has been a common complaint about matplotlib. However, the __pyplot.style__ package allows the use of pre-defined stylesheets.

Use `plt.style.available` to see a list of stylesheets available to your system. These are the ones included in matplotlib:

```python
print(plt.style.available)
```

```python
plt.style.use('fivethirtyeight')
plt.plot(data)
```
