```python
%matplotlib inline  
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 5, 10)
y = x ** 2
```

```python
plt.plot(x,y, 'r')
plt.xlabel('x')
plt.ylabel('y')
plt.title('The title')
```

```python
plt.subplot(1,2,1)
plt.plot(x, y, 'r--')
plt.subplot(1,2,2)
plt.plot(y, x, 'g*-')
```

```python
fig, axes = plt.subplots()
axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('TTT')
```
